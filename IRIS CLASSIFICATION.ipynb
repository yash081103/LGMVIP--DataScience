### IMPORTING LIBRARIES ESSENTIAL FOR THE PROJECT


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### LOADING DATASET

```python
col=["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width","Species"]
iris = pd.read_csv("iris.data",names=col)
iris.head()
```

### DIVIDING DATASET INTO DEPENDENT AND INDEPENDENT VARIABLES

```python
x=iris.iloc[:,0:-1].values
y=iris.iloc[:,-1].values
```

### SPLITTING DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

## MODEL SELECTION

### MODEL1: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred1=model.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred1)*100)
```

### MODEL2:K-nearest neighbours

```python
#k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(x_train,y_train)
```

```python
pred2=model2.predict(x_test)
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_test,pred2))
```

### MODEL3: DecisisonTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
```

```python
pred3=model3.predict(x_test)
print("accuracy:",accuracy_score(y_test,pred3))
for i in range(len(pred3)):
    print(y_test[i],pred3[i])
```

### CLASSIFICATION REPORT

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred3))
```
